#构建损失函数computer(X,y,w)

```python
def computeCost(X,y,w):
    inner=np.power(((X*w.T)-y),2)
    return np.sum(inner)/(2*len(X))
```

#在训练集中插入一列，以方便计算代价和梯度

```python
data.insert(0,'Ones',1)
data
```

```python
cols=data.shape[1]
X=data.iloc[:,:cols-1]#X是所有行，去掉最后一列
y=data.iloc[:,cols-1:]#y是所有行，最后一列
```

```python
X.head()
```

```python
y.head()
```

代价函数应该是numpy矩阵，所以我们需要转换X和Y，然后才能使用它们，我们还需要初始化w

```python
X=np.matrix(X.values)
y=np.matrix(y.values)
w=np.matrix(np.array([0,0]))
w
```

看下维度

```python
X.shape,w.shape,y.shape
```

计算代价函数（theta初始值为0）

```python
computeCost(X,y,w)
```

```python
def batch_gradientDescent(X,y,w,alpha,iters):
    temp=np.matrix(np.zeros(w.shape))
    parameters=int(w.ravel().shape[1])
    cost=np.zeros(iters)
    
    for i in range(iters):
        error=(X*w.T)-y
        
        for j in range(parameters):
            term=np.multiply(error,X[:,j])
            temp[0,j]=w[0,j]-((alpha/len(X))*np.sum(term))
        w=temp
        cost[i]=computeCost(X,y,w)
        
    return w,cost
```

初始化一些附加变量-学习率α和要执行的迭代次数

```python
alpha=0.01
iters=30000
```

现在我们运行梯度下降算法来将我们的参数θ适合于训练集

```python
g,cost=batch_gradientDescent(X,y,w,alpha,iters)
g
```

最后，我们可以使用我们拟合的参数计算训练模型的代价函数（误差）

```python
computeCost(X,y,g)
```

现在我们来绘制线性模型和数据，直观地看出它的拟合

```python
x=np.linspace(data['天销量/千克'].min(),data['天销量/千克'].max(),100)
f=g[0,0]+(g[0,1]*x)

fig,ax=plt.subplots(figsize=(12,8))
ax.plot(x,f,'r',label='预测值')
ax.scatter(data['天销量/千克'],data['天总加成率'],label='训练数据')
ax.legend(loc=2)
ax.set_xlabel('天销量/千克率',fontsize=10)
ax.set_ylabel('天总加成',rotation=0,fontsize=18)
ax.set_title('预测天销量/千克和天总加成率',fontsize=18)
plt.show()
```

我们也可以快速查看这一个训练进程

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(np.arange(iters),cost,'r')
ax.set_xlabel('迭代次数',fontsize=18)
ax.set_ylabel('代价',rotation=0,fontsize=18)
ax.set_title('误差和训练Epoch数',fontsize=18)
plt.show()
```

# 多变量线性回归分析

```python
data2=pd.read_excel('茄子20.xlsx',sheet_name='Sheet3')
data2.head()
```

对于此任务，我们添加了另一个预处理步骤-特征归一化，这个对于pandas来说很简单

```python
data2=(data2-data2.mean())/data2.std()
data2.head()
```

```python
#add ones column
data2.insert(0,'Ones',1)

#set X(training data)and y (target variable)
cols=data2.shape[1]
X2=data2.iloc[:,0:cols-1]
y2=data2.iloc[:,cols-1:cols]

#convert to matrices and initialize theta
X2=np.matrix(X2.values)
y2=np.matrix(y2.values)
w2=np.matrix(np.array([0,0,0]))

#perform linear regression on the data set
g2,cost2=batch_gradientDescent(X2,y2,w2,alpha,iters)

#get the cost(error)of the model
g2,computeCost(X2,y2,g2)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(np.arange(iters),cost2,'r')
ax.set_xlabel('迭代次数',fontsize=18)
ax.set_ylabel('代价',rotation=0,fontsize=18)
ax.set_title('误差和训练Epoch数',fontsize=18)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
x=np.asarray(X[:,1])
y=np.asarray(y)
model.fit(x,y)
f1=model.predict(x).flatten()
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(x,f1,'r',label='预测值LR')
ax.scatter(data['天销量/千克'],data['天总加成率'],label='训练数据')
ax.legend(loc=2)
ax.set_xlabel('天销量/千克',fontsize=10)
ax.set_ylabel('天总加成',rotation=0,fontsize=18)
ax.set_title('预测天销量/千克和天总加成率',fontsize=18)
plt.show()
```

# L1正则化

```python
from sklearn.linear_model import Lasso
model=Lasso(alpha=0.001)
x=np.asarray(X[:,1])
y=np.asarray(y)
model.fit(x,y)
f3=model.predict(x).flatten()
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(x,f3,'r',label='预测值Lasso')
ax.scatter(data['天销量/千克'],data['天总加成率'],label='训练数据')
ax.legend(loc=2)
ax.set_xlabel('天销量/千克',fontsize=10)
ax.set_ylabel('天总加成',rotation=0,fontsize=18)
ax.set_title('预测天销量/千克和天总加成率',fontsize=18)
plt.show()
```

# 调参

```python
from sklearn.model_selection import cross_val_score
alphas=np.logspace(-3,-2,50)
test_scores=[]
for alpha in alphas:
    clf=Lasso(alpha)
    test_score=np.sqrt(-cross_val_score(clf,x,y,cv=5,scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
import matplotlib.pyplot as plt
plt.plot(alphas,test_scores)
plt.title('Alpha vs CV Error')
plt.show()
```
